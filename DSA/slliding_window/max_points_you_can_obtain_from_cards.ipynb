### DESCRIPTION (inspired by Leetcode.com)

Given an array of integers representing card values, write a function to calculate the maximum score you can achieve by picking exactly k cards.

You must pick cards in order from either end. You can take some cards from the beginning, then switch to taking cards from the end, but you cannot skip cards or pick from the middle.

For example, with k = 3:

Take the first 3 cards: valid
Take the last 3 cards: valid
Take the first card, then the last 2 cards: valid
Take the first 2 cards, then the last card: valid
Take card at index 0, skip some, then take card at index 5: not valid (skipping cards)

Constraints: 1 <= k <= cards.length

Example 1: 

Input:

cards = [2,11,4,5,3,9,2]
k = 3

Output:

17

Explanation:

First 3 cards: 2 + 11 + 4 = 17
Last 3 cards: 3 + 9 + 2 = 14
First 1 + last 2: 2 + 9 + 2 = 13
First 2 + last 1: 2 + 11 + 2 = 15
Maximum score is 17.

Example 2: 

Input:

cards = [1, 100, 10, 0, 4, 5, 6]
k = 3

Output:

111

Explanation: Take the first three cards: 1 + 100 + 10 = 111. This is better than taking the last 3 cards (4 + 5 + 6 = 15) or any other combination.

```python
class Solution_V1:
    def maxScore(self, cards: list[int], k: int) -> int:
        if len(cards) == k:
            return sum(cards)
        
        window_sum = 0
        for left in range(k):
            window_sum += cards[left]

        max_sum = window_sum
        for right in range(len(cards)-1, len(cards)-1-k, -1):
            window_sum -= cards[left] 
            window_sum += cards[right]
            left -= 1        
            max_sum = max(max_sum, window_sum)

        return max_sum
```

### FEEDBACK
Good job — your solution is correct and passes all tests.

A couple of things to tighten up for interview-quality code:

You’re relying on left from the first loop inside the second loop. That works in Python, but it’s a bit fragile and harder to read because left is reused as an index-like variable.
The logic is a valid sliding-window idea, but it would be clearer if you explicitly tracked the number of cards taken from the left and right, or used a window over the middle n - k cards.
Watch out for readability: window_sum actually represents the sum of the currently chosen cards, not a window in the usual sense.
Overall: correct approach, just make the state update easier to follow.

```python
class Solution:
    def maxScore(self, cards: list[int], k: int) -> int:
        total_score = sum(cards)
        if len(cards) == k:
            return total_score
        
        start = 0
        window_len = len(cards) - k
        window_score = 0
        max_score = float('-inf')

        for end in range(len(cards)):
            window_score += cards[end]

            if end-start+1 == window_len:
                max_score = max(max_score, total_score - window_score)
                window_score -= cards[start]
                start += 1

        return max_score
```

```python
from dataclasses import dataclass
from typing import Callable

@dataclass(frozen=True)
class Input:
    nums: list[int]
    k: int

@dataclass(frozen=True)
class Test:
    input: Input
    expected_result: int

def run_tests(tests: list[Test], func: Callable[[list[int], int], int]):
    for test in tests:
        result = func(test.input.nums, test.input.k)
        if result == test.expected_result:
            print("Test passed for " + str(test.input.nums))
        else:
            print(f"Test failed for {test.input.nums}. Expected: {test.expected_result}, Actual: {result}")
```

### FEEDBACK
Good job — your solution is correct and passes all tests.

You used the key insight well: instead of choosing k cards directly, you maximize the sum of the remaining n-k contiguous cards. That gives an O(n) sliding-window solution.

A couple of small improvements:

You can avoid float('-inf') here, since the window always exists when k < len(cards); initializing max_score = 0 is simpler.
Consider naming unpicked_cards_score as window_sum for clarity.
Your code already handles the k == len(cards) edge case cleanly.
