Given an array heights where each element represents the height of a vertical line, pick two lines to form a container. Return the maximum area (amount of water) the container can hold.

What is area? Width × height, where width is the distance between walls, and height is the shorter wall (water overflows at the shorter wall).

```python
class Solution:
    def max_area(self, heights: list[int]) -> int:
        if len(heights) < 2:
            return -1

        left, right = 0, len(heights)-1
        current_max = 0

        while left < right:
            height = min(heights[left], heights[right])
            width = right - left
            current_area = height*width
            current_max = max(current_max, current_area)

            if heights[left] < heights[right]:
                left += 1
            else:
                right -= 1

        return current_max
```

```python
from dataclasses import dataclass
from typing import Callable

@dataclass(frozen=True)
class Test:
    input: list
    expected_output: int

def run_tests(tests: list[Test], func: Callable[[list], int]):
    for test in tests:
        result = func(test.input)
        if result == test.expected_output:
            print("Test passed for " + str(test.input))
        else:
            print(f"Test failed for {test.input}. Expected: {test.expected_output}, Actual: {result}")
```

```python
tests = [Test(input = [1,8,6,2,5,4,8,3,7], expected_output = 49),
         Test([1,1], 1),
         Test([4,3,2,1,4], 16),
         Test([1,2,1], 2),
         Test([3,4,1,2,2,4,1,3,2], 21),
         Test([1,100,100,1], 100),
         Test([5,0,5], 10)]
run_tests(tests, Solution().max_area)
```
