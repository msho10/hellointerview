### DESCRIPTION (inspired by Leetcode.com)

Write a function to sort a given integer array nums in-place (and without the built-in sort function), where the array contains n integers that are either 0, 1, and 2 and represent the colors red, white, and blue. Arrange the objects so that same-colored ones are adjacent, in the order of red, white, and blue (0, 1, 2).

```python
class Solution_V1:
    def sortColors(self, nums: list[int]) -> None:
        left, right = -1, len(nums)
        if len(nums) < 2:
            return
        
        def swap(m, n):
            temp = nums[m]
            nums[m] = nums[n]
            nums[n] = temp

        i = 0
        while i < right:
            if nums[i] == 0:
                left += 1
                swap(left, i)
                if i == left:
                    i += 1
            elif nums[i] == 2:
                right -= 1
                swap(right, i)
            else:
                i += 1
```

Nice work — your solution is correct and passes all tests. You implemented the Dutch national flag partitioning in-place with O(n) time and O(1) extra space.

A couple of interview-level notes:

if i == left: i += 1 is a bit fragile/unclear. It works here, but it’s easy to reason about incorrectly. In this pattern, it’s usually cleaner to advance i whenever you place a 0, and not advance it when you swap in a 2 so the new value gets rechecked.
swap() is fine, but Python tuple swap is shorter and idiomatic.
Your early return for small arrays is okay.
Overall: solid solution; just make the pointer movement rules easier to follow.

```python
class Solution:
    def sortColors(self, nums: list[int]) -> None:
        left, right = 0, len(nums)-1
        if len(nums) < 2:
            return

        i = 0
        while i <= right:
            if nums[i] == 0:
                #since the only possible values for nums[left] here is 0 or 1, 
                #there is no need to check the value of nums[i] after the swap, 
                #and thus OK to increment i here.
                nums[left], nums[i] = nums[i], nums[left] 
                left += 1
                i += 1  
            elif nums[i] == 2:
                # don't increment i here, we need to check the new nums[i] value 
                # after the swap which could be any value.
                nums[right], nums[i] = nums[i], nums[right] 
                right -= 1
            else:
                i += 1
```

```python
from dataclasses import dataclass
from typing import Callable

@dataclass(frozen=True)
class Test:
    input: list[int]
    expected_result: list[int]

def run_tests(tests: list[Test], func: Callable[[list[int]], None]):
    for test in tests:
        nums = test.input.copy()
        func(nums)
        if nums == test.expected_result:
            print("Test passed for " + str(test.input))
        else:
            print(f"Test failed for {test.input}. Expected: {test.expected_result}, Actual: {nums}")
```
