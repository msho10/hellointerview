### Description

Given an array of integers nums and an integer k, find the maximum sum of any contiguous subarray of size k.

Example 1: 

Input:

nums = [2, 1, 5, 1, 3, 2]
k = 3

Output:

9

Explanation: The subarray with the maximum sum is [5, 1, 3] with a sum of 9.

```python
class Solution:
    def maxSum(self, nums: list[int], k: int) -> int:
        if len(nums)<k or k == 0:
            return 0
        
        max_sum_ = float('-inf')
        current_sum = 0
        start = 0

        for end in range(len(nums)):
            current_sum += nums[end]
            
            if end-start+1 == k:    # check the window size against k
                max_sum_ = max(max_sum_, current_sum)
                    
                # slide window
                current_sum -= nums[start]
                start += 1

        return max_sum_
```

### Feedback

A couple of interview-level notes:

Your handling of len(nums) < k and k == 0 is sensible, but check the problem constraints: if k is guaranteed valid, that guard may be unnecessary.
Since max_sum_ starts at -inf, you correctly handle all-negative arrays too.
Your window logic is clean and runs in O(n) time, O(1) extra space.
Minor readability improvement: you could use more descriptive names like window_sum and avoid the trailing underscore unless needed to avoid a clash.

```python
from dataclasses import dataclass
from typing import Callable

@dataclass(frozen=True)
class Input:
    nums: list[int]
    k: int

@dataclass(frozen=True)
class Test:
    input: Input
    expected_result: int

def run_tests(tests: list[Test], func: Callable[[list[int], int], int]):
    for test in tests:
        result = func(test.input.nums, test.input.k)
        if result == test.expected_result:
            print("Test passed for " + str(test.input.nums))
        else:
            print(f"Test failed for {test.input.nums}. Expected: {test.expected_result}, Actual: {result}")
```
