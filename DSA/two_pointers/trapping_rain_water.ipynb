### DESCRIPTION (inspired by Leetcode.com)
Write a function to calculate the total amount of water trapped between bars on an elevation map, where each bar's width is 1. The input is given as an array of n non-negative integers height representing the height of each bar.

### MY SOLUTION
First find the left and right wall that can trap water. 
A left wall is found when the current wall is taller than the wall to its right. 
A right wall is found when current wall is taller than the wall to its left.

When 2 walls are found, use a third pointer to iterate the walls between them. 
The amount of water is the difference between the shorter of the 2 walls and the current wall. 
However, if current wall is taller than any of the 2 walls, that means we have 2 separate traps. 
Use recussion to calculate water for each trap by using the current wall as left wall for one 
and as right wall for the other. 

```python
class Solution:
    def trappingWater_V1(self, height: list[int]) -> int:
        i, j = 0, len(height)-1
        left, right = -1, -1

        # find the trapping walls left and right
        while i < j:
            if left == -1 and height[i] > height[i+1]:
                left = i
            if right == -1 and height[j] > height[j-1]:
                right = j
            if left > -1 and right > -1:
                return self._cal_water(height, left, right)
            i += 1
            j -= 1        
        
        return 0

    def _cal_water(self, height: list[int], left: int, right: int) -> int:
        water = 0
        short_height = height[left] if height[left] < height[right] else height[right]
        
        for i in range(left+1, right):
            if height[i] <= short_height:
                water += short_height - height[i]
            else:
                return self._cal_water(height, left, i) + self._cal_water(height, i, right)
            
        return water
```

A few interview-level concerns:

Worst-case complexity can degrade: _cal_water() may recurse repeatedly when it hits a higher bar in the middle, which can lead to O(n²) behavior on some patterns.
Stack risk: deep recursion can hit Python’s recursion limit for large inputs (n մինչև 20,000).
Edge handling is brittle: height[i+1] / height[j-1] assumes enough room; your outer loop avoids some cases, but the logic is still harder to reason about than necessary.
In an interview, a two-pointer + running max approach is safer and guarantees O(n) time, O(1) space.
