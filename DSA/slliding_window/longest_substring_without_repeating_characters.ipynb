# DESCRIPTION (inspired by Leetcode.com)
Write a function to return the length of the longest substring in a provided string s where all characters in the substring are distinct.

Example 1: 

Input:

s = "eghghhgg"
Output:

3

The longest substring without repeating characters is "egh" with length of 3.

Example 2: 

Input:

s = "substring"

Output:

8

The answer is "ubstring" with length of 8.

```python
class Solution:
    def longestSubstringWithoutRepeat(self, s: str) -> int:
        start = 0
        max_len = 0
        freq_map = {}

        for end in range(len(s)):
            freq_map[s[end]] = freq_map.get(s[end],0) + 1
            
            while(freq_map[s[end]] > 1):
                freq_map[s[start]] -= 1
                if freq_map[s[start]] == 0:
                    del freq_map[s[start]]
                start += 1
           
            max_len = max(max_len, end-start+1)  
            
        return max_len
```

For readability, add spaces after commas: freq_map.get(s[end], 0).
Edge cases like "", all unique chars, and all repeated chars are handled correctly.



```python
from dataclasses import dataclass
from typing import Callable

@dataclass(frozen=True)
class Test:
    input: str
    expected_result: int

def run_tests(tests: list[Test], func: Callable[[str], int]):
    for test in tests:
        result = func(test.input)
        if result == test.expected_result:
            print("Test passed for " + str(test.input))
        else:
            print(f"Test failed for {test.input}. Expected: {test.expected_result}, Actual: {result}")
```

```python
tests = [
    Test('', 0),
    Test('a', 1),
    Test('aa', 1),
    Test('abc', 3),
    Test('aaa', 1),
    Test('aabcdefgg', 7),
    Test('eghghhgg', 3),
    Test('substring', 8),
    Test('This is a test', 5),
]

run_tests(tests, Solution().longestSubstringWithoutRepeat)
```
