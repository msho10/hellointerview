Given an integer array nums, write a function to rearrange the array by moving all zeros to the end while keeping the order of non-zero elements unchanged. Perform this operation in-place without creating a copy of the array.

```python
class Solution_v1:
    def moveZeroes(self, nums: list[int]) -> None:
        if len(nums) < 2: return
        left, right = 0, 1
        
        while True:
            #left is a moving pointer and will stop at zeros
            while nums[left] != 0:
                left += 1
                if left >= len(nums) - 1:
                    return

            #right is ahead of the left pointer and will stop at nonzero numbers
            right = left + 1
            while nums[right] == 0:
                right += 1
                if right >= len(nums):
                    return

            if left < right:
                nums[left] = nums[right]
                nums[right] = 0
            else:
                return

            
```

Nice job — your solution passes and preserves order.

A few things to watch for in interviews:

Your code is correct on the tested cases, but the pointer logic is a bit fragile. The nested while True plus multiple early returns makes it harder to reason about boundary conditions.
There’s a subtle risk of indexing past the end in the inner loops if the array ends with zeros or has no non-zero after left; you guard most cases, but this is easy to get wrong under pressure.
You can make this cleaner and still in-place with a single pass and one write pointer, which is easier to prove correct and has the same O(n) time / O(1) space.
In interviews, favor the simplest invariant: “all non-zeros before nextNonZero are already placed.”

```python
class Solution_v2:
    def moveZeroes(self, nums: list[int]) -> None:
        nextNonZero = 0

        for i in range(len(nums)):
            if nums[i] != 0:
                nums[nextNonZero], nums[i] = nums[i], nums[nextNonZero]
                nextNonZero += 1
```

```python
from dataclasses import dataclass
from typing import Callable

@dataclass
class Test:
    original_list: list[int]
    expected_result: list[int]

def run_tests(tests: list[Test], func: Callable[[list[int]], None]):
    for test in tests:
        input = test.original_list.copy()
        func(input)
        if input == test.expected_result:
            print("Test passed for " + str(test.original_list))
        else:
            print(f"Test failed for {test.original_list}. Expected: {test.expected_result}, Actual: {input}")
```

```python
tests = [
    Test([2,0,4,0,9], [2,4,9,0,0]),
    Test([2,3,4,0,9], [2,3,4,9,0]),
    Test([2], [2]),
    Test([0], [0]),
    Test([2,0], [2,0]),
    Test([0,2], [2,0]),
    Test([2,1,0], [2,1,0]),
    Test([0,1,2], [1,2,0]),
    Test([1,0,2], [1,2,0]),
    Test([1,1,1,1], [1,1,1,1]),
    Test([0,0,0,0], [0,0,0,0]),
    Test([1,0,0,1], [1,1,0,0]),
]
run_tests(tests, Solution_v2().moveZeroes)
```
