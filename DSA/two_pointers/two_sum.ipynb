Given a sorted array of integers nums, determine if there exists a pair of numbers that sum to a given target.

```python
class Solution:
    def twoSum(self, nums: list[int], target: int) -> bool:
        if len(nums) < 2: return False

        left, right = 0, len(nums)-1
        while left < right:
            current_sum = nums[left] + nums[right]

            if current_sum == target: 
                return True
            
            if current_sum < target:
                left += 1
            else:
                right -= 1

        return False
```

```python
from dataclasses import dataclass
from typing import Callable

@dataclass(frozen=True)
class Test:
    input: list
    target: int
    expected_output: bool

def run_tests(tests: list[Test], func: Callable[[list], bool]):
    for test in tests:
        result = func(test.input, test.target)
        if result == test.expected_output:
            print("Test passed for " + str(test.input))
        else:
            print(f"Test failed for {test.input}. Expected: {test.expected_output}, Actual: {result}")
```

```python
tests = [Test([1,3,4,6,8,10,13], 13, True),
         Test([1,3,4,6,8,10,13], 6, False),
         Test([], 0, False),
         Test([2], 6, False),
         Test([6], 6, False),
         Test([3,3], 6, True),
         Test([2, 3], 6, False),
         Test([-1, -6], -7, True),
         Test([-10, 16], 6, True),
         Test([1,2,3,4,5], 10, False),
         Test([2,3,5,7,11], 9, True),
         Test([1,1,3,4,6], 2, True),
         Test([1,2147483646], 2147483647, True),
         Test([0,2147483647], 2147483647, True),
         Test([1000000000,1000000000], 2000000000, True)
]
run_tests(tests, Solution().twoSum)
```
