Write a function to count the number of triplets in an integer array nums that could form the sides of a triangle.

For three sides to form a valid triangle, all three of these conditions must hold: (a + b > c), (a + c > b), and (b + c > a), where (a), (b), and (c) are the side lengths. In other words, the sum of every possible pair must exceed the third side.

```python
class Solution:
    def triangleNumber(self, nums: list[int]) -> int:
        if len(nums) < 3:
            return 0
        
        count = 0
        nums.sort()
        for a in range(len(nums)-2):
            for b in range(a+1, len(nums)-1):
                for c in range(b+1, len(nums)):
                    if nums[a] + nums[b] > nums[c]:
                        count += 1
        return count
```

```python
from dataclasses import dataclass
from typing import Callable

@dataclass(frozen=True)
class Test:
    input: list[int]
    expected_output: int

def run_tests(tests: list[Test], func: Callable[[list[int]], int]):
    for test in tests:
        result = func(test.input)
        if result == test.expected_output:
            print("Test passed for " + str(test.input))
        else:
            print(f"Test failed for {test.input}. Expected: {test.expected_output}, Actual: {result}")
```

A couple of interview-level improvements:

Your approach is O(n^3) because of the three nested loops. That works for small inputs, but it will time out on larger arrays.
Since you already sort nums, you can exploit the ordering to reduce this to O(n^2) with a two-pointer sweep.
Minor cleanup: you can return early for len(nums) < 3, which you already did well.
For coding interviews, mentioning the complexity tradeoff is important even when the brute-force solution is accepted.

```python
tests = [Test([11,4,9,6,15,18],10),
         Test([0],0),
         Test([1,1],0),
         Test([0,0,0],0),
         Test([1,1,1],1),
         Test([1,1,2],0),
         ]
run_tests(tests, Solution().triangleNumber)
```
